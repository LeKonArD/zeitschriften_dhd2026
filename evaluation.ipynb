```python
def clean(x):
    r = []
    for n in x:
        if n.startswith("k"):
            r.append("k")
            continue
        if n.startswith("p"):
            r.append("p")
            continue
        if n.startswith("an") or n.startswith("f"):
            r.append("a")
            continue
        if n.startswith("am") or n.startswith("n") or n.startswith("g"):
            r.append("n")
            continue
        print(n)
    return r
```

```python
data = pd.read_csv("evaldata/haltung.tsv", sep="\t")

tab = []
for m in set(data["model"]):
    s = []
    for r in [1,2,3,4,5]:

        s.append(f1_score(clean(list(data[(data.model == m) & (data.run == r)].true)), list(clean(data[(data.model == m) & (data.run == r)].Haltung)),  average="weighted"))

    tab.append([m, np.mean(s), np.std(s)])

tab = pd.DataFrame(tab, columns=["model","mean","std"]).sort_values("mean", ascending=False)
tab.sort_values("mean", ascending=False)
```
